```python
import numpy as np
import pandas as pd
import nltk
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()
```

```python
raw_data = pd.read_csv("train.csv")
train = raw_data.copy()
train.describe(include='all')
```

```python
train.iloc[7612]
```

```python
def fun(word):
    return wordnet_lemmatizer.lemmatize(word, pos='v')
train['keyword'] = train.keyword[train['keyword'].notnull()].apply(fun)
train.head()
```

```python
unique_keyword1 = np.array(train.keyword[train.keyword.notnull()].unique())
len(unique_keyword1)
```

```python
train['text'] = train['text'].apply(nltk.word_tokenize)
```

```python
train_lem = train.copy()
train_lem
```

```python
def fun(sentence_words):
    punctuations="?:!.,;#[({]})/@!$%^&*"
    s = []
    for word in sentence_words:
        if word not in punctuations:
            w = word.split()
            t = []
            for x in w:
                if x not in punctuations:
                    t.append(x)
            t = t.join()
            s.append(wordnet_lemmatizer.lemmatize(t,pos='v'))
    return (s)
train_lem['text'] = train_lem['text'].apply(fun) 
```

```python
train_lem.describe(include='all')
```

```python
train_lem = train_lem.drop(['location'], axis=1)
```

```python
train_lem.dropna(subset=['keyword']).describe(include='all')
```

```python
data_preprocessed = train_lem.copy()
```

```python
x_train = data_preprocessed.drop(['id','target'], axis=1)
y_train = data_preprocessed['target']
```

```python
x_train.text = x_train.text.apply(' '.join)
x_train
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
token_vector = vectorizer.fit_transform(x_train.text)
```

```python
token_vector.shape
```

```python
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB().fit(token_vector, y_train)
```

```python
test = pd.read_csv('test.csv')
test.describe(include='all')
```

```python
test = test.drop(['keyword','location'], axis=1)
```

```python
test['text'] = test['text'].apply(nltk.word_tokenize)
test
```

```python
def fun(sentence_words):
    punctuations="?:!.,;#[({]})/@!$%^&*"
    s = []
    for word in sentence_words:
        if word not in punctuations:
            s.append(wordnet_lemmatizer.lemmatize(word,pos='v'))
    return (s)
test['text'] = test['text'].apply(fun) 
test
```

```python
test.text = test.text.apply(' '.join)
test
```

```python
test_vectors = vectorizer.transform(test.text)
test_vectors.shape
```

```python
predicted = clf.predict(test_vectors)
predicted
```

```python
test['target'] = predicted
test
```

```python
test = test.drop(['text'],axis=1)
```
